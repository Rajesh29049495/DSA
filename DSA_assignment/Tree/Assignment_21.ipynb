You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

## function to do the traversal and print the values of nodes of the tree into an array        
def inorder_traversal(node, array):
    if node is None:
        return
    inorder_traversal(node.left, array)
    array.append(node.value)
    inorder_traversal(node.right, array)

    
## function to insert value in the tree
def insert(node, value):
    if node is None:
        return TreeNode(value)
    
    if value<node.value:
        node.left= insert(node.left, value)
    else:
        node.right = insert(node.right, value)
    return node

## main function which first call the function to traverse the binary tree and fill the array with values of the tree and then call teh function to insert the vlaues in the BST form
def constructBST(root):
    arr=[]
    inorder_traversal(root, arr)
    
    if not arr:
        return None
    node = None
    for val in arr:
        node = insert(node, val)
    return node

## function to check if the formed tree is BST
def isValidBST(node):
    l =[]
    inorder_traversal(node, l)
    
    for i in range(1, len(l)):
        if l[i]<l[i-1]:
            return False
    return True
    

# Construct the binary tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

root_bst = constructBST(root)

print("checking if the formed tree is BST:", isValidBST(root_bst))


## function to print the tree using inorder traversal
def printBST(node):
    if node is None:
        return
    printBST(node.left)
    print(node.value)
    printBST(node.right)

printBST(root_bst)
```

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

node-1 = 6

**Output-1:**

The distance between the two keys = 4

**Input-2:**

**Output-2:**

The distance between the two keys = 2

```python
def distance(root, p, q):
    if root is None:
        return None
    
    if root.value==p or root.value ==q:
        return 0
    
    left_distance = distance(root.left, p, q)
    right_distance = distance(root.right, p, q)
    
    if left_distance is not None and right_distance is not None:
        # Both nodes are found in different subtrees, so the distance is the sum of distances from teh root to each node
        return left_distance + right_distance +2
    elif left_distance is not None:
        # when one of teh node found in left subtree and other at the root node at that instance
        return left_distance + 1
    elif right_distance is not None:
        # when one of teh node found in right subtree and other at the root node at that instance
        return right_distance + 1
    
    # Both nodes not found in the current subtree
    return None

# Construct the binary tree
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

result = distance(root, 6, 14)
print("The distance between the two keys:", result)
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

## main function which first locate the node w.r.t. the two values between distanec we have to find, then it call teh function to locate the LCA of teh two nodes then call teh function to count distanec between LCA and each node then finally add them and return
def find_distance(root, node1_value, node2_value):
    # Find the nodes in the BST
    n1 = find_node(root, node1_value)
    n2 = find_node(root, node2_value)

    # Find the lowest common ancestor
    lca = find_lca(root, n1, n2)

    # Calculate the distance from LCA to both nodes
    dist1 = find_distance_to_node(lca, n1, 0)
    dist2 = find_distance_to_node(lca, n2, 0)

    # Return the distance between the two nodes
    return dist1 + dist2

def find_node(root, value):
    if root is None or root.value == value:
        return root

    if value < root.value:
        return find_node(root.left, value)
    else:
        return find_node(root.right, value)

def find_lca(root, node1, node2):
    if root is None or node1 is None or node2 is None:
        return None

    if root.value > node1.value and root.value > node2.value:
        return find_lca(root.left, node1, node2)

    if root.value < node1.value and root.value < node2.value:
        return find_lca(root.right, node1, node2)

    return root

def find_distance_to_node(root, target, distance):
    if root is None:
        return 0

    if root.value == target.value:
        return distance

    if target.value < root.value:
        return find_distance_to_node(root.left, target, distance + 1)
    else:
        return find_distance_to_node(root.right, target, distance + 1)
    

# Construct the binary tree
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

result = find_distance(root, 6, 14)
print("The distance between the two keys:", result)
```

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>

```python
class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def convert_to_doubly_linked_list(root):
    # Base case: if root is None, return None
    if root is None:
        return None
    
    # Convert the left subtree to a doubly linked list
    left_list = convert_to_doubly_linked_list(root.left)
    
    # Create a new doubly linked list node for the current root
    new_node = DoublyLinkedListNode(root.value)
    
    # Connect the left_list to the new_node
    if left_list is not None:
        left_tail = left_list
        while left_tail.next is not None:
            left_tail = left_tail.next
        left_tail.next = new_node
        new_node.prev = left_tail
    
    # Convert the right subtree to a doubly linked list
    right_list = convert_to_doubly_linked_list(root.right)
    
    # Connect the new_node to the right_list
    if right_list is not None:
        right_head = right_list
        right_head.prev = new_node
        new_node.next = right_head
    
    # Return the head of the doubly linked list
    if left_list is not None:
        return left_list
    else:
        return new_node

    
# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert the binary tree to a doubly linked list
head = convert_to_doubly_linked_list(root)
```

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

4 â†’ 5

5 â†’ 6

6 â†’ 7

7 â†’ -1

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None  # Pointer to the next node at the same level


def connect_nodes_at_same_level(root):
    # Base case: if root is None, return
    if root is None:
        return
    
    # Create a queue for level-order traversal
    queue = []
    queue.append(root)
    
    # Perform level-order traversal
    while queue:
        level_size = len(queue)
        prev_node = None
        
        for _ in range(level_size):
            current_node = queue.pop(0)
            
            # Connect the current node to the previous node at the same level
            if prev_node is not None:
                prev_node.next = current_node
            
            prev_node = current_node
            
            # Add the children of the current node to the queue
            if current_node.left is not None:
                queue.append(current_node.left)
            if current_node.right is not None:
                queue.append(current_node.right)
        
        # Connect the last node in the current level to None
        prev_node.next = None

        
# Create the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Connect nodes at the same level
connect_nodes_at_same_level(root)
```
