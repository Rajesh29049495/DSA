Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.


</aside>

````python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

## function to do the traversal and print the values of nodes of the tree into an array        
def inorder_traversal(node, array):
    if node is None:
        return
    inorder_traversal(node.left, array)
    array.append(node.value)
    inorder_traversal(node.right, array)

    
## function to construct a  DLL from an array    
def binary_tree_to_dll(root):
    arr = []
    
    inorder_traversal(root, arr)
    
    if not arr:
        return None

    head = Node(arr[0])  # Create the first node
    prev = head

    for i in range(1, len(arr)):
        node = Node(arr[i])  # Create a new node for each element
        prev.next = node
        node.prev = prev
        prev = node

    return head
````

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

Input:
```
        6
     /    \
    3      5
  /   \     \
 2     5     4
     /   \
    7     4
```
Output:

There are 4 leaves, hence 4 root to leaf paths -
  6->3->2
  6->3->5->7
  6->3->5->4
  6->5>4

</aside>

````python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def print_root_to_leaf_paths(root):
    if root is None:
        return

    stack = [(root, str(root.value))]  # Initialize the stack with the root node and its path

    while stack:
        node, path = stack.pop()

        if node.left is None and node.right is None:
            print(path)  # Print the path when a leaf node is encountered
        if node.right:
            stack.append((node.right, path + "->" + str(node.right.value)))  # Push right child to the stack with updated path
        if node.left:
            stack.append((node.left, path + "->" + str(node.left.value)))  # Push left child to the stack with updated path

# Create the binary tree
root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)
root.right.right = TreeNode(4)

# Print root-to-leaf paths
print_root_to_leaf_paths(root)
````

GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

**Examples:**

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 2 4 5 3
        Postorder -> 4 5 2 3 1
Output : 

Yes
Explanation : 

All of the above three traversals are of
the same tree 

                           1
                         /   \
                        2     3
                      /   \
                     4     5

Input : 

        Inorder -> 4 2 5 1 3
        Preorder -> 1 5 4 2 3
        Postorder -> 4 1 2 3 5
Output : 

No

</aside>

````python
def check_same_tree(preorder, inorder, postorder):
    if not preorder or not inorder or not postorder:
        return True

    if len(preorder) != len(inorder) or len(inorder) != len(postorder):
        return False

    root = preorder[0]
    root_index = inorder.index(root)

    inorder_left = inorder[:root_index]
    inorder_right = inorder[root_index + 1:]

    preorder_left = preorder[1:root_index + 1]
    preorder_right = preorder[root_index + 1:]

    postorder_left = postorder[:root_index]
    postorder_right = postorder[root_index:-1]

    return (
        check_same_tree(preorder_left, inorder_left, postorder_left) and
        check_same_tree(preorder_right, inorder_right, postorder_right)
    )

# Test case 1
inorder = [4, 2, 5, 1, 3]
preorder = [1, 2, 4, 5, 3]
postorder = [4, 5, 2, 3, 1]
print(check_same_tree(preorder, inorder, postorder))  # Output: True

# Test case 2
inorder = [4, 2, 5, 1, 3]
preorder = [1, 5, 4, 2, 3]
postorder = [4, 1, 2, 3, 5]
print(check_same_tree(preorder, inorder, postorder))  # Output: False
````
